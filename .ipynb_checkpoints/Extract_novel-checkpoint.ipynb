```python
import jieba

with open("./stopwords.txt",'r') as stop:
    stopwords = stop.read()
    print(stopwords)

with open("./test.txt") as f:
    text = f.read()
    paras = text.split(' ')
    pool = list()
for para in paras:
    words = jieba.cut(para)
    for word in words:
        if word not in stopwords:
            pool.append(word)
#             print(word)
print(pool)
type(pool)
```

```python
dict={}
for term in pool:
    if term in dict:
        dict[term]=dict[term]+1
    else:
        dict[term]=1
print(dict)
```

```python
import operator
sorted_x = sorted(dict.items(), key=operator.itemgetter(1), reverse=True)
print(sorted_x[0:5])
```
