```python
import sqlite3
import datetime

import ptt_crawler
```

```python
sqlite3.version
```

```python
# 連結資料庫，若無則建新檔
conn = sqlite3.connect('./test.sqlite3')
```

```python
c = conn.cursor()
type(c)
```

```python
# 建立資料庫column name & type
c.execute("""CREATE TABLE article
             (url VARCHAR(255) PRIMARY KEY,
              title VARCHAR(255),
              content TEXT,
              board VARCHAR(255),
              author VARCHAR(255),
              ip VARCHAR(255),
              dt TIMESTAMP)""")
```

```python
conn.in_transaction
```

```python
# 先測試只爬 one page
ptt_articles = ptt_crawler.crawl_ptt(1)
```

```python
# 查看一下抓下來的資料
ptt_articles
```

```python
c.executemany("""INSERT OR REPLACE INTO article VALUES (:url, :title, :content, :board, :author, :ip, :dt)""", ptt_articles)
```

```python
conn.in_transaction
```
