```python
import pyLDAvis
import pyLDAvis.sklearn
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
pyLDAvis.enable_notebook()
```

```python
import csv

with open('../test1.csv', 'r') as f:
    reader = csv.reader(f, delimiter=',')
    mydoc = list(tuple(rec) for rec in reader)
```

```python
docraws=[]
for rec in mydoc:
    docraws.append(rec[1])
```

```python
docraws
```

```python
vect = CountVectorizer()
```

```python
myMatrix = vect.fit_transform(docraws)
```

```python
myMatrix.shape
```

```python
LDA = LatentDirichletAllocation(n_topics=10, random_state=0)
```

```python
LDA.fit(myMatrix)
```

```python
pyLDAvis.sklearn.prepare(LDA, myMatrix, vect, mds='mmds')
```

```python
pyLDAvis.sklearn.prepare(LDA, myMatrix, vect, mds='tsne')
```

```python
def print_top_words(model, feature_names, n_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print("Topic #%d:" % topic_idx)
        print(" ".join([feature_names[i] for i in topic.argsort()[:-n_top_words - 1:-1]]))
    print()
```

```python
n_top_words = 15
```

```python
print_top_words(LDA, vect.get_feature_names(), n_top_words)
```
